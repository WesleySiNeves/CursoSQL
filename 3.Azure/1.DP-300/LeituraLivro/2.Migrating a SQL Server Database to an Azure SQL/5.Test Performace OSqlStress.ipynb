# <mark>\## Ostress config</mark>

```powershell
ostress.exe -E -dTreinamentoSQL -Q"SELECT YEAR([Data]) as Ano, MONTH([Data]) as Mes ,TotalCredito= SUM(Valor) from Lancamentos
GROUP BY YEAR([Data]),MONTH([Data])" -n30 -r10 -q
```
